```python
X = dataset.iloc[:, 1:]
Y = dataset.iloc[:, 0]
```

```python
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3, stratify=Y)
```

```python
print(X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train_scaled = StandardScaler().fit_transform(X_train)
X_test_scaled = StandardScaler().fit_transform(X_test)

X_train = X_train_scaled.copy()
X_test = X_test_scaled.copy()

print(X_train[0].mean(), X_train[0].std())
```

```python
print(f"Train Set:\n{pd.Series(Y_train).value_counts()}\n\n")
print(f"Test Set:\n{pd.Series(Y_test).value_counts()}\n\n")

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
sns.histplot(Y_train, bins = len(np.unique(Y_train)))

plt.subplot(1, 2, 2)
sns.histplot(Y_test, bins = len(np.unique(Y_test)))

plt.tight_layout()
plt.show()
```

```python
import imblearn
```

```python
smote = imblearn.over_sampling.SMOTE(random_state=3, k_neighbors=3)

X_train_resample, Y_train_resample = smote.fit_resample(X_train, Y_train)

print(X_train_resample.shape, Y_train_resample.shape)
```

```python
print(pd.Series(Y_train_resample).value_counts())
```

```python
X_train, Y_train = X_train_resample.copy(), Y_train_resample.copy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
avg = lambda x: x.sum()/len(x)
models = [LogisticRegression(max_iter=1000), SVC(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier(n_neighbors=3)]
model_names = ["Logistic Regression", "SVM", "Decision Tree", "Random Forest", "KNN"]
accuracy = []

for idx, model in enumerate(models):
    accuracy.append(avg(cross_val_score(model, X_train, Y_train, cv=5)))
    print(f"{model_names[idx]} trained")

model_accuracy = {x:y for x, y in zip(model_names, accuracy)}
print(model_accuracy)
```

```python
model = models[np.argmax(accuracy)]
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import base
```

```python
parameters = {
    "C":[0.1, 1, 10, 100],
    "max_iter":[100, 1000, 10000],
    "solver":["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}

random_model = GridSearchCV(base.clone(model), param_grid=parameters, cv=5)
random_model.fit(X_train, Y_train)
```

```python
print(f"Best Parameters: {random_model.best_params_}")

pd.DataFrame(random_model.cv_results_)
```

```python
model = random_model.best_estimator_
```

```python
print(model.score(X_test, Y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_test_pred = model.predict(X_test)

conf_matrix = confusion_matrix(Y_test, y_test_pred)
sns.heatmap(conf_matrix, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
clf_report = classification_report(Y_test, y_test_pred)
print(clf_report)
```
