```python
X = dataset.drop(columns=['Diet', 'Exercises'])
Y = dataset['Exercises']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnb = RandomForestClassifier(n_estimators=100, random_state=42)
rnb.fit(X_train, Y_train)
```

```python
rnb.score(X_test, Y_test)*100, rnb.score(X_train, Y_train)*100
```

```python
rnb.predict([[1,18,1.68,47.5,0,1]])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_test_pred = rnb.predict(X_test)

conf_matrix = confusion_matrix(Y_test, y_test_pred)
sns.heatmap(conf_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
clf_report = classification_report(Y_test, y_test_pred)
print(clf_report)
```

```python
X2 = dataset.drop(columns=['Diet', 'Exercises'])
Y2 = dataset['Diet']
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, test_size=0.2, random_state=42)
```

```python
rnb2 = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rnb2.fit(X2_train, Y2_train)
```

```python
rnb2.score(X2_test, Y2_test)*100, rnb2.score(X2_train, Y2_train)*100
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
avg = lambda x: x.sum()/len(x)
models = [LogisticRegression(max_iter=1000), SVC(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier(n_neighbors=3)]
model_names = ["Logistic Regression", "SVM", "Decision Tree", "Random Forest", "KNN"]
accuracy = []

for idx, model in enumerate(models):
    accuracy.append(avg(cross_val_score(model, X2_train, Y2_train, cv=5)))
    print(f"{model_names[idx]} trained")

model_accuracy = {x:y for x, y in zip(model_names, accuracy)}
print(model_accuracy)
```

```python
import numpy as np
```

```python
model = models[np.argmax(accuracy)]
model.fit(X2_train, Y2_train)
```

```python
model.score(X2_test, Y2_test)*100, model.score(X2_train, Y2_train)*100
```

```python
y2_test_pred = model.predict(X2_test)

conf_matrix = confusion_matrix(Y2_test, y2_test_pred)
sns.heatmap(conf_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
clf_report = classification_report(Y2_test, y2_test_pred)
print(clf_report)
```
